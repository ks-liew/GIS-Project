```python
import requests
import json
import re
from pyquery import PyQuery 
import numpy as np
import networkx as nx

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def hasNumbers(inputString):
    return any(char.isdigit() for char in inputString)
```

```python
lines = []
stations = []

for u in ['U1', 'U2', 'U3', 'U4', 'U6']:
    overpass_url = "http://overpass-api.de/api/sketch-line?operator=Wiener+Linien&style=wuppertal&width=1300&ref=" + u
    response = requests.get(overpass_url)
    m = re.findall('\>.*?\<', response.text)

    res = []
    for i, el in enumerate(m):
        if i < 13 or el == '><' or len(el) < 7 or 'WEST' in el or hasNumbers(el):
            continue
        res.append(el[1:-1])
    stations += res
    lines.append(np.array(res))
```

```python
print(len(np.unique(stations)))
stations_integers = list(range(0, 97))
```

```python
np.unique(stations)
```

```python
toIntegers = dict(zip(np.unique(stations), stations_integers))
toString = dict(zip(stations_integers, np.unique(stations)))
```

```python
toIntegers.keys()
```

```python
graph_matrix = np.zeros((97,97))

for l in lines:
    for i in range(0, len(l) - 1):
        graph_matrix[toIntegers[l[i]], toIntegers[l[i+1]]] += 1
        graph_matrix[toIntegers[l[i+1]], toIntegers[l[i]]] += 1
        
```

```python
graph_matrix[0]
```

```python
lines_encoded = []
for line in lines:
    line_encoded = []
    for i, el in enumerate(line):
        line_encoded.append(toIntegers[el])
    lines_encoded.append(line_encoded)
lines_encoded
```

```python
G = nx.from_numpy_matrix(graph_matrix)

pos = nx.kamada_kawai_layout(G)
plt.figure(1,figsize=(150,150))
nx.draw_networkx_nodes(G,pos, nodelist=lines_encoded[0],node_color='#d60a22', node_size=15000,font_size=65)
nx.draw_networkx_nodes(G,pos, nodelist=lines_encoded[1],node_color='#b30ad6', node_size=15000,font_size=65)
nx.draw_networkx_nodes(G,pos, nodelist=lines_encoded[2],node_color='#fca905', node_size=15000,font_size=65)
nx.draw_networkx_nodes(G,pos, nodelist=lines_encoded[3],node_color='#23a521', node_size=15000,font_size=65)
nx.draw_networkx_nodes(G,pos, nodelist=lines_encoded[4],node_color='#a56e20', node_size=15000,font_size=65)
# nx.draw_networkx(G, pos,node_size=5000,font_size=30, labels=toString)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos, labels=toString)
plt.show()
```
